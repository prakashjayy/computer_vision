## Topic-1 Brief introduction to datasets.

Most of the usecases in Deep learning is fueled by large scale datasets. Be it ImageNet dataset for image classification, Openimagesv4 dataset for object detection, COCO, ADE20k for segmentation or MegaFace dataset for face recognition, we have a large scale dataset for each and every topic. Over the last few years deep learning community have spent considerable resources in collecting, analysing and annotating these datasets as there was a direct correlation between the performance of deep learning models to the dataset. the larger the dataset, the larger the peformance. 

Segmentation datasets require annotating each and every pixel on the image. This is obviously very difficult compared to image classification annotation (assigning a label to the image) and object detection annotation (Drawing a rectangular box on the image). So though we don't see datasets with million images like imagenet, There are considerably large datasets available for research now and the entire deep learning field is concentrating on this now as segmentation is the ultimate form of converting image (unstructured) data to structured data. 

Also if you are solving any image problem in you organization, the first and foremost thing we need to keep in mind is that we need to annotate large scale dataset. One way to get away with this in computer vision is finding network architecutures and pre-trained weights availability. This will help us in solving our problem by annotating less data than required. Now if we are considering image segmentation as a solution to your problem, we need to know what all datasets are available and later check if any research team has trained any models on these datasets and released pre-trained weights. This is the first fundamental research you as a data science practioner has to do. 

The segmentation community has various public datasets available. Some of the noted datasets are **ADE20k**, **CityScapes** and **COCO** datasets. We will look at these datasets in detail in the next section.


### 1.1 ADE20K
ADE20K is the largest open source dataset for semantic segmentation and scene parsing, released by MIT Computer Vision team. The organization has a [website](http://groups.csail.mit.edu/vision/datasets/ADE20K/) and also a [research paper](https://arxiv.org/pdf/1608.05442.pdf) explaining the reason and process of collecting datasets.

Some of the key features of the dataset are as follows
- After annotation, there are 20,210 images in the training set, 2,000 images in the validation set, and 3,000 images in the testing set.
- There are in total 3,169 class labels annotated, among them 2,693 are object and stuff classes, 476 are object part classes. All the images are exhaustively annotated with objects. Many objects are also annotated with their parts. For each object there is additional information about whether it is occluded or cropped, and other attributes. Sample images and annotations from the ADE20K dataset are show below.

![ADE20K dataset](../images/ade20k.png)

The annotations on the dataset (as shown in the above diagram) is done at 3 levels. We can see in the below section a relation tree of objects and parts for the dataset. Each number indicates the number of instances for each object. The full relation tree is available at the dataset [webpage](http://groups.csail.mit.edu/vision/datasets/ADE20K/). First object is annotated (For example Oven). Then parts of the of object (oven) are annotated ( Example button panel and door) and then under button panel we have buttons, dial and screen. Like wise every scene is annotated by expert annotator.

![Section tree](../images/ade20k_section_tree.png)

The organization has a [website](https://www.cityscapes-dataset.com/) where you can find any information of the data.
- The organization also released a [paper](https://arxiv.org/abs/1604.01685) on the same to explain the way they have collected, annotated and dealt with the dataset. They also explained the important charateristics of the dataset and why it is important to collect this kind of information.

- assessing the performance of vision algorithms for two major tasks of semantic urban scene understanding: pixel-level and instance-level semantic labeling;

### 1.3 Coco Dataset 
Common Objects in Context aka AKA is a large scale object detection, segmentation and captioning dataset. coco has several features. The organization has also released a paper https://arxiv.org/pdf/1405.0312.pdf. 
There are two kind of categories.
  - stuff: includes materials and objects with no clear boundaries. Example: Sky, street, grass.
  - thing: objects for which individual instances can be easily labelled. Example: Person, car, cat etc.
- For stuff level annotations they have released another paper https://arxiv.org/pdf/1612.03716.pdf.
    
<img src="http://calvin.inf.ed.ac.uk/wp-content/uploads/data/cocostuffdataset/cocostuff-examples.png" alt="COCO-Stuff example annotations" width="100%">

#### Highlights of the dataset
- 164K complex images from COCO
- Dense pixel-level annotations
- 80 thing classes, 91 stuff classes and 1 class 'unlabeled'
- Instance-level annotations for things from COCO
- Complex spatial context between stuff and things
- 5 captions per image from COCO 

The hirearicy of the things and stuff objects is as follows  
![coco_label and stuff hierarcy]("../images/cocostufflabelhierarchy.png")

To be compatible with COCO, **COCO-Stuff** has 91 thing classes (1-91), 91 stuff classes (92-182) and 1 class "unlabeled" (0). Note that 11 of the thing classes of COCO do not have any segmentation annotations (blender, desk, door, eye glasses, hair brush, hat, mirror, plate, shoe, street sign, window). The classes desk, door, mirror and window could be either stuff or things and therefore occur in both COCO and COCO-Stuff. To avoid confusion we add the suffix "-stuff" or "-other" to those classes in COCO-Stuff. The full list of classes and their descriptions can be found [here](https://github.com/nightrome/cocostuff/blob/master/labels.md).

It is often quite important to check the annotations visually to understand if everything going as expected or not. In pycocotools there are several functions which allow you to do this. In this section lets try and visualize any image along with its annoations using img_id. The process is as follows.

- For collecting the annotations of the image_id. We have **cocoSegmentationToSegmentationMap** function in **cocostuffhelper** module.
- visualize the image and mask using our helper function written below.

````python
## Helper tools to visualize the images
import matplotlib.pyplot as plt
%matplotlib inline

def plot_img_and_mask(img, mask):
    """ Plots the image and mask using matplotlib.
    """
    fig, axs = plt.subplots(ncols=2, figsize=(10, 5), sharex=True, sharey=True)
    axs[0].imshow(img)
    axs[1].imshow(mask)
    fig.tight_layout()
    plt.show()
````

````python
from pycocotools.cocostuffhelper import cocoSegmentationToSegmentationMap
from utils import plot_img_and_mask
import urllib.request
from PIL import Image

## Selecting the image_id
imgId = 122851

## Location to save the data 
save_loc = "../../datasets/"

## Collecting the image info
img_info = data.imgs[imgId]

## If you don't have the train_data, download the image using the url. Default is true
if True:
    img = urllib.request.urlretrieve(img_info["coco_url"], save_loc+img_info["file_name"])

## Load the image
image= Image.open(save_loc+img_info["file_name"])

## load the mask for the image
mask = cocoSegmentationToSegmentationMap(data,
                                 imgId=imgId)

## Plot the image
plot_img_and_mask(image, mask)
````

## 3.2 Count Occurance of each label (pixel count) in the dataset.
Not all classes are represented in the equal number in a dataset. With stuff objects this will be more evident as **sky**, **grass** usually occupy more space in a image then **cow** , **dog**. Usually for any dataset be it image or not it is very important to know the imbalance in the dataset, so that we can later take some algorithmic decisions in tackling this or verify our models accuracy class wise and check if underrepresented classes are facing any difficulty in detection.

We can process each and every annotation and collect all the pixel labels to check imbalance but this will take a lot of time as there are more than 1.18L images in the dataset. For this we will randomly select a few 1000 images and process them and plot. This stat will give an initial glance at the data stats, we can increase the number of images to get more accurate stats. 

Below is the code to do that analysis

````python
from tqdm import tqdm ## To track the progress
import numpy as np

## list all the image_ids
imgids = list(data.imgs.keys())

## Total image_ids 
total_img_ids = len(imgids)

## Randomly select 1000 images
sample_imgids = [imgids[np.random.randint(total_img_ids)] for i in range(1000)]

## Initializing an empty dictionory
class_counter = {}

## Looping through all the images
for imgid in tqdm(sample_imgids):
    mask = cocoSegmentationToSegmentationMap(data,
                                 imgId=imgid)
    for pixel in mask.reshape(-1):
        if pixel not in class_counter.keys():
            class_counter[pixel] = 1 
        else:
            class_counter[pixel] += 1
````

````python
## List all class numbers
classes_list = list(class_counter.keys())

## List the count of each class on the sample data obtained.\
## If statement is used too avoid all pixels which doesn't contain any labels.
class_name_count = [class_counter[i] for i in classes_list if i != 0]

## List all class names from numbers
classes_name = [data.cats[int(i)]["name"] for i in classes_list if i != 0]

## Sort by occurances 
sorted_classes_name = [classes_name[i] for i in np.argsort(class_name_count)]
sorted_class_name_count = [class_name_count[i] for i in np.argsort(class_name_count)]

## Plot using matplotlib
plt.figure(figsize=(8, 18))
plt.barh(sorted_classes_name, sorted_class_name_count)
plt.show()
````

## Some of the observations we can infer from the graph are 
- There is a heavy imbalance in the dataset.
- The other category in the above belongs to all the thing objects annotated in coco. we need to use instances_train2017.json files to get those annotations.
- Common elements like sky, trees and grass are present the most.

## Topic-4: Different types of segmentation using pycocotools

We have already analysed the stuff annotations. We will now use the same methods to load the thing annotations. Note that in stuff annotations all the object annotations are masked and kept in other classes.
To understand the different types of segementation and how they are processed using pycocotools, Lets load both stuff annotations and thing annotations together.

````python
## loading thing objects annotations using coco
thing_data = coco.COCO("../../data/cocostuff/annotations/instances_train2017.json")
stuff_data = coco.COCO("../../data/cocostuff/annotations/stuff_train2017.json")
````

We need to collect all the image_ids which have 3-5 objects in the image. This is done to make sure that we have multiple instances of objects so that we can easily find out the difference between semantic, instance and panoptic segmentation.

````python
## List image_ids
image_ids = list(thing_data.imgs.keys())

## List of image ids which have 3-5 objects in the image
req_img_id = []
for img in image_ids:
    anns_id = thing_data.getAnnIds(img)
    if len(anns_id) <= 5 and len(anns_id) >2 :
        req_img_id.append(img)
print("Total image_ids with 3-5 objects in the image: {}".format(len(req_img_id)))
````

````python
## select an image_id randomly. Here i am choosing 100th image_id in the requried img_id 
imgid = req_img_id[100]

## Collect all the anns 
anns_id = thing_data.getAnnIds(imgid)
anns = [thing_data.anns[i] for i in anns_id]

## Find the mask of each thing label
masks = [thing_data.annToMask(i) for i in anns]

## Collect all the thing labels 
labels = [i["category_id"] for i in anns]

## Collect all the thing labels name 
label_names = [thing_data.cats[i]["name"] for i in labels]

## Collect all the thing boxes 
boxes = [i["bbox"] for i in anns]

## Collect the segmentation mask of stuff objects. we are using data instead of thing_data
stuff_mask = cocoSegmentationToSegmentationMap(stuff_data,
                                               imgId=imgid)

## Collecting the image info
img_info = stuff_data.imgs[imgid]

## If you don't have the train_data, download the image using the url. Default is true
if True:
    img = urllib.request.urlretrieve(img_info["coco_url"], save_loc+img_info["file_name"])
image= Image.open(save_loc+img_info["file_name"])
````

````python
## Height and width of the image
height = img_info["height"]
width = img_info["width"]
````

- Background objects like sky, trees etc (stuff) are not recogniszed.
- Each individual instance of any object is separately detected (Two humans in the below diagram have indvidual instances), whereas in semantic segmentation the network directly doesn't separates the objects.
- Unlike object detection, we can segment each object pixel wise
